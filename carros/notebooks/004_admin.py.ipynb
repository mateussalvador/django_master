# Admin do django
- É através do painel do admin do django que o admin-user pode criar coisas no banco de dados, deletar, etc.
- Para que seja possível acessar essa página, é necessário inicializar o banco de dados.

## Inicializando o banco de dados
- Usamos os comandos `python manage.py makemigrations` e o `python manage.py migrate`
- Pode ser feito antes mesmo de criar models nos nossos apps, visto que o django possui suas próprias tabelas necessárias para o seu funcionamento. 

### makemigrations
- Analisa o projeto intiero e monta as migrations. São arquivos de código python que possuem comandos para o banco de dados (para criar bancos de dados, para apagar, etc)

### migrate
- Executa as migrations, ou seja, executa os comandos no banco de dados de fato.
- Ao rodar esses comandos, com o uso da extensão SQLite Viewer já é possível visualizar as tabelas padrões do django. Existe uma tabela chamada `auth_user` que é onde os usuários do django são cadastrados, inclusive os super usuários.

# Criando um super usuário para conseguir acessar o admin do django
- Para criar um superuser no django, rodamos o comando `python manage.py createsuperuser`
- Após isso, já podemos acessar o `admin/` no nosso servidor.
- Ao criar essa primeira vez, só teremos os administradores de grupos e de usuários que são padrões do django, visto que ainda não criamos nossas models.

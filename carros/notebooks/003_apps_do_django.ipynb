# Apps
- Os apps do django so divises lgicas da nossa aplicao, podendo ser feita conforme nossa necessidade.
- Por exemplo, podemos criar uma aplicao para carros, responsvel por todo o gerenciamento de registro de carros (criar carros, deletar, update, etc).
- Tambm podemos ter uma app para os vendedores, onde gerenciaremos o cadastro de vendedores. 

## Como selecionar apps?
- Podemos colocar todo o nosso projeto dentro de um s app, bem como poderamos ter tudo dentro da pasta que carrega as configuraes (corao). Porm, o mais correto  dividir de forma lgica: **Uma app deve gerenciar uma coisa, uma entidade, deve ser responsvel pelo gerenciamento da lgica de uma parte especfica**. Isso facilita inclusive na manuteno.

# Criando a primeira app
- Para criar um app, usamos o comando `python manage.py startapp nome_app`. O nome do nosso app ser `cars`.
- Ao execut-lo, ser criada uma pasta com o nome do app com uma srie de novos arquivos.

## Observao importante sobre o app
- Precisamos informar ao django que ele precisa usar essa app que acaba de ser criada.
- Para isso, vamos em `settings.py` e acrescentamos o nome da app ao fim de `INSTALLED_APPS`. Caso contrrio, o django ir ignorar essa app.

---

# Arquivos da app
- A pasta do app traz os seguintes arquivos:

## `tests.py`
-  um arquivo usado para escrever testes

## `apps.py`
- Traz configuraes do app, como o nome e outras. Podemos adicionar mais configuraes conforme necessidades mais avanadas.

## `models.py`
- Usado em todas as apps.  aqui criamos as models (tabelas do banco de dados) com seus respectivos campos e tipos de dados.

## `views.py`
- Tambm  muito importante, to quando as models.
- Definem a lgica da aplicao a ser executada quando um usurio acessar uma determinada rota, mostrando/renderizando o contedo correto.

## `admin.py`
- Configura o painel adminitrativo do django. Essas pginas so acessveis apenas aos administradores do sistema.
- Atravs do admin conseguimos um CRUD completo das entidades.

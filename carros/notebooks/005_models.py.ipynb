- Vamos iniciar o desenvolvimento da tabela de carros para que possamos cadastrar, excluir e gerenciar o registro de carros.
- Inicialmente, tudo será desenvolvido na visão de administrador do sistema, não necessariamente de usuários comuns (que verão num template, por exemplo).

## Criando o arquivo
- Em `models.py`, temos que o arquivo deve iniciar com `from django.db import models`
```python

from django.db import models

class Car(models.Model): # Deve herdar da classe Model, que já vem pronta do django
    # Abaixo escrevemos os campos das tabelas
    id = models.AutoField(primary_key=True) # Cria um campo que se autoincrementa e determina que esse campo será a chave primária
    model = models.CharField(max_length=200) # modelo do carro. CharField precisa de max_length, que é o tamanho máximo do texto (colocamos 200 caracteres). Blank True significa que ao criar um carro novo, podemos deixar o texto em branco e também podemos aceitar o valor null. Também são campos que podemos colocar e recebem os valores True ou False 
    brand = models.CharField(max_length=200) # marca do carro
    factory_year = models.IntegerField(blank=True, null=True) # ano de fabricação. Vamos armazenar somente o ano, então pode ser um inteiro
    model_year = models.IntegerField(blank=True, null=True) #ano de modelo
    value = models.DecimalField(blank=True, null=True) #Uma opção para substituir o FloatField, visto que é mais preciso.

```

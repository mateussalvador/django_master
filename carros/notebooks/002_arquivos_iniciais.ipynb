# Arquivos iniciais
- Ao iniciar um projeto, o django cria os seguintes arquivos: o `manage.py` e uma pasta com o nome do projeto, contendo: `__init.py`, `asgi.py`, `settings.py`, `urls.py`, `wsgi.py`.

## `manage.py`
- Sempre usaremos este arquivo para rodar arquivos.
- Ele tem os mesmos comandos que o `django-admin`, mas importa as configurações do nosso projeto. Ou seja, ele sabe onde encontrar as configurações do nosso projeto, sabendo o escopo, banco de dados, entre outras, sabendo lidar com o nosso projeto.
- Normalmente, o `django-admin` só é usado para criar o projeto.

## `settings.py`
- Fica dentro do coração do nosso projeto e armazena todas as configurações do nosso projeto. 

## `__init__.py`
- Diz ao interpretador que aquele diretório é um pacote.

## `wsgi.py` e `asgi.py`
- Gerencia as configurações do nosso servidor de aplicação quando vamos colocar o projeto em produção.

## `urls.py`
- É onde configuramos as urls do nosso projeto.

## `db.sqlite3`
- É um banco de dados que já vem configurado no django para usarmos no ambiente local de desenvolvimento.

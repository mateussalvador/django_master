# Primeiro projeto
- Construíremos um projeto inicial mais direcionado à venda de carros para entendermos os conceitos iniciais do Django.
- Para isso, iremos criar uma pasta raiz chamada carros para elaborar este projeto.
- Para iniciar o projeto, criamos um ambiente virtual para instalarmos as dependências do projeto (`pip install django`)
- Para verificar se o django foi instalado corretamente, rodamos o `django-admin --version`

---

# django-admin
- Ao instalar o Django, ele cria um atalho no terminal chamado `django-admin` que possui alguns comandos para gerenciar o projeto.
- O primeiro é o `django-admin startproject nome_projeto .` para criar um projeto. Ele criará uma pasta com o nome do nosso projeto.
- O nome do nosso projeto será `app`. Essa pasta será o coração do nosso projeto. 
- Para verificar se tudo está funcionando, rodamos o `python manage.py runserver numero_porta`. Por padrão, roda na porta 8000. Esse comando roda o servidor de desenvolvimento que vem embutido no django, permitindo colocar nossa aplicação no ar dentro da nossa máquina local.

## O que é o `manage.py`?
- É um arquivo que roda comandos internos dentro da aplicação django.
